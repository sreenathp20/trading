```python
from connect import *
from datetime import datetime
start = datetime(2023, 3, 1)
end = datetime(2023, 3, 2)
u = Upstox()
df1 = u.getDfData('nifty50', start, end)
```

```python
df1.head()
```

```python

import seaborn as sns

import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(figsize=(15, 10))
 
d2 = df1[['date','close',  'ema_close']]

#plot sales of each store with custom colors
sns.lineplot(x='date', y='value', hue='variable', 
             data=pd.melt(d2, ['date']),
             palette=['red', 'blue', 'purple', 'pink'], ax=ax)

plt.show()
```

```python
df1['ema_close_p2'] = df1['close'].ewm(alpha=0.2).mean()
df1['ema_close_p1'] = df1['close'].ewm(alpha=0.1).mean()
df1['ema_close_p3'] = df1['close'].ewm(alpha=0.3).mean()


```

```python
df1
```

```python
fig, ax = plt.subplots(figsize=(15, 10))
 
d2 = df1[['date','close', 'ema_close_p3']]

#plot sales of each store with custom colors
sns.lineplot(x='date', y='value', hue='variable', 
             data=pd.melt(d2, ['date']),
             palette=['red', 'blue', 'purple', 'pink'], ax=ax)

plt.show()
```

```python
df1['ema_close_p3_sma3'] = df1['ema_close_p3'].rolling(3).mean()
```

```python
fig, ax = plt.subplots(figsize=(15, 10))
 
d2 = df1[['date','close', 'ema_close_p3_sma3']]

#plot sales of each store with custom colors
sns.lineplot(x='date', y='value', hue='variable', 
             data=pd.melt(d2, ['date']),
             palette=['red', 'blue', 'purple', 'pink'], ax=ax)

plt.show()
```
